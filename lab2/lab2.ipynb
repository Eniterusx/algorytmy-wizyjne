```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
min_area = 40
bin_threshold = 10
file_name = "pedestrian"

f = open("data/%s/temporalROI.txt" % file_name, "r")
line = f.readline()
roi_start, roi_end = line.split()
roi_start = int(roi_start)
roi_end = int(roi_end)

start_frame = roi_start
end_frame = roi_end
step_size = 2
```

```python
frame_before = cv2.imread("data/%s/input/in%06d.jpg" % (file_name, start_frame))
frame_before_gray = cv2.cvtColor(frame_before, cv2.COLOR_BGR2GRAY)

TP = 0
TN = 0
FP = 0
FN = 0

for i in range(start_frame, end_frame, step_size):
    frame = cv2.imread("data/%s/input/in%06d.jpg" % (file_name, i))
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    mask = cv2.imread("data/%s/groundtruth/gt%06d.png" % (file_name, i))
    mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)

    difference = cv2.absdiff(frame_gray, frame_before_gray)
    (T, bin_image) = cv2.threshold(difference, bin_threshold, 255, cv2.THRESH_BINARY)

    # Erode and dilatate
    kernel = np.ones((3, 3), np.uint8)
    for _ in range(5):
        bin_image = cv2.erode(bin_image, kernel, iterations=1)
        bin_image = cv2.dilate(bin_image, kernel, iterations=1)

    retval, labels, stats, centroids = cv2.connectedComponentsWithStats(bin_image)

    I_VIS = frame
    if stats.shape[0] > 1:
        tab = stats[1:, 4]
        pi = np.argmax(tab) + 1
        # tab = np.sort(tab)
        # pi = 1
        # while pi < stats.shape[0] and stats[pi, 4] > min_area:
        if stats[pi, 4] > min_area:
            cv2.rectangle(I_VIS,(stats[pi, 0],stats[pi, 1]),(stats[pi, 0]+stats[pi, 2],stats[pi, 1]+stats[pi, 3]),(0,0,150),2)
            cv2.putText(I_VIS, "%f" % stats[pi, 4], (stats[pi, 0], stats[pi, 1]), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,150))
            cv2.putText(I_VIS, "%d" % pi, (int(centroids[pi, 0]), int(centroids[pi, 1])), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,150))
            # pi += 1

    TP += np.logical_and((mask == 255), (labels > 0)).sum()
    TN += np.logical_and((mask == 0), (labels == 0)).sum()
    FP += np.logical_and((mask == 0), (labels > 0)).sum()
    FN += np.logical_and((mask == 255), (labels == 0)).sum()



    # cv2.imshow("labels", np.uint8(labels / retval * 255))
    cv2.imshow("I_VIS", I_VIS)
    cv2.waitKey(35)

    frame_before_gray = frame_gray
cv2.destroyAllWindows()

precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)
print("Precision: %f" % precision)
print("Recall: %f" % recall)
print("F1: %f" % f1)
```
